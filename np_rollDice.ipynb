```python
#Simulate rolling two dice 1000 times using np.random.randint()

import numpy as np

# Simulate rolling two dice 1000 times
rolls = np.random.randint(1, 7, size=(1000, 2))

# Calculate the sum of each roll
sums = np.sum(rolls, axis=1)

# Print the first 10 rolls
print("First 10 rolls:")
print(rolls[:10])
print("\nFirst 10 sums:")
print(sums[:10])
```

```python
import numpy as np

# Simulate rolling two dice 1000 times
dice1 = np.random.randint(1, 7, size=1000)
dice2 = np.random.randint(1, 7, size=1000)

# Sum of both dice
sum_of_dice = dice1 + dice2

# Print first 10 results
print("First 10 rolls of Dice 1:", dice1[:10])
print("First 10 rolls of Dice 2:", dice2[:10])
print("First 10 sums:", sum_of_dice[:10])
```
