```python
library = {}

def add_book(title, author, genre, year):
    if title in library:
        print("Book already exists.")
    else:
        library[title] = {"Author": author, "Genre": genre, "Year": year}
        print("Book added.")

def remove_book(title):
    print("Book removed." if library.pop(title, None) else "Book not found.")

def search_books(criteria, value):
    return {a:b for a,b in library.items() if b.get(criteria) == value}

def update_book(title, criteria, value):
    if title in library and criteria in library[title]:
        library[title][criteria] = value
        print("Book updated.")
    else:
        print("Invalid title or field.")

def generate_report():
    if not library:
        print("Library is empty.")
        return
    for a,b in library.items():
        print(f"\nTitle: {a}, Author: {b['Author']}, Genre: {b['Genre']}, Year: {b['Year']}")

def calculate_statistics():
    genres = {}; authors = {}; years = {}
    for b in library.values():
        genres[b['Genre']] = genres.get(b['Genre'], 0) + 1
        authors[b['Author']] = authors.get(b['Author'], 0) + 1
        years[b['Year']] = years.get(b['Year'], 0) + 1
    print("\nTotal Books:", len(library))
    print("Genres:", genres)
    print("Authors:", authors)
    print("Years:", years)

while True:
    print("\n1.Add  2.Remove  3.Search  4.Update  5.Report  6.Stats  7.Quit")
    c = input("Choice: ")

    if c == "1":
        add_book(input("Title: "), input("Author: "), input("Genre: "), int(input("Year: ")))

    elif c == "2":
        remove_book(input("Title: "))

    elif c == "3":
        crit = input("Criteria (Author/Genre/Year): ").capitalize()
        val = input("Value: ")
        val = int(val) if crit == "Year" else val
        for a,b in search_books(crit, val).items():
            print(f"\nTitle: {a}, Author: {b['Author']}, Genre: {b['Genre']}, Year: {b['Year']}")
        if not search_books(crit, val):
            print("No books found.")

    elif c == "4":
        crit = input("Field (Author/Genre/Year): ").capitalize()
        val = input("New value: ")
        val = int(val) if crit == "Year" else val
        update_book(input("Title: "), crit, val)

    elif c == "5":
        generate_report()

    elif c == "6":
        calculate_statistics()

    elif c == "7":
        break

    else:
        print("Invalid choice.")
```
