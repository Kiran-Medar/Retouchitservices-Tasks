```python
print("===  USERNAME CREATOR ===")

# Step 1: Collect user details
fname = input("What's your first name? ").strip()
lname = input("What's your last name? ").strip()
Sentence = input("Type your sentence: ").strip()

# Step 2: Generate creative usernames using string tricks

# Style 1: Take first 3 letters of each name (lowercase)
option1 = fname[:3].lower() + lname[:3].lower()

# Style 2: Join names in title case with a dash
option2 = "-".join([fname.title(), lname.title()])

# Style 3: Reverse first name and add first letter of last name (uppercase)
option3 = fname[::-1] + lname[0].upper()

# Style 4: Use f-string with word and total length
option4 = f"{fname.lower()}_{sentence.upper()}_{len(fname) + len(lname)}"

# Step 3: Show all generated usernames
print("\n Here are some cool username ideas for you:")
print(f"1️ {option1}")
print(f"2️ {option2}")
print(f"3️ {option3}")
print(f"4️ {option4}")

# Step 4: Let user select their favourite
pick = input("\nSelect your favourite username (1-4): ").strip()

# Step 5: Show the final username
if pick == "1":
    final_name = option1
elif pick == "2":
    final_name = option2
elif pick == "3":
    final_name = option3
elif pick == "4":
    final_name = option4
else:
    final_name = "Invalid selection"

print(f"\n Your final username is: {final_name}")
```
