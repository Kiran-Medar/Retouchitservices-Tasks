```python
#Build a program to manage student grades using lists and tuples
students = ["john", "george" ,"riya" ,"Alice"]
grades = [8, 9, 6, 7]

print(students,grades)
students.append("william")
students.append("kilia")
print(students)

grades.append(5)
grades.append(9)
print(grades)

print(f"first grade of the list: ",grades[1])
print(f"Third grade of the list: ",grades[3])

print("\n--- Student Grades ---")
for student, grade in zip(students, grades):
    print(f"student {student} - Grade: {grade}")
```
