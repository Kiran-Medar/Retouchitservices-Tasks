```python
class A:
    def __init__(self):
        print("initializing the A")

    def foo(self):
        pass

class B(A):
    def __init__(self):
        super().__init__()
        print("initializing the B")

    def foo(self):
        return "Implementation of B"

class C(A):
    def __init__(self):
        super().__init__()
        print("initializing the C")

    def foo(self):
        return "Implementation of C"

class D(B, C):
    def __init__(self):
        super().__init__()
        print("initialising the  D")

    def foo(self):
        b = B.foo(self) ## call B s foo to avoid ambiguity
        c = C.foo(self) ## call c s foo to avoid ambiguity
        print(f"D resolves diamond by using: ({b}) AND ({c})")


## Main
d = D()
d.foo()
```
