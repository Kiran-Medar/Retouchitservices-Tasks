```python
# STUDENT MANAGEMENT SYSTEM using Dictionaries and Sets

# Dictionary to store all student records
students = {
    101: {"name": "Alice Johnson", "age": 20,
          "courses": {"Math", "Physics", "Chemistry"}, "gpa": 3.8},

    102: {"name": "Rob Smith", "age": 22,
          "courses": {"Computer Science", "Math", "English"}, "gpa": 3.5},

    103: {"name": "Charlie Brown", "age": 21,
          "courses": {"Physics", "Biology", "Chemistry"}, "gpa": 3.9}
}

# Set of all available courses in the college
all_courses = {
    "Math", "Physics", "Chemistry", "Biology",
    "Computer Science", "English", "History"
}

print("\n=== STUDENT MANAGEMENT SYSTEM ===\n")


# Functions

def add_student(roll, name, age, gpa, courses):
    if roll in students:
        print("Student already exists!")
    else:
        students[roll] = {
            "name": name,
            "age": age,
            "gpa": gpa,
            "courses": set(courses)
        }
        print(f"Student '{name}' added successfully")


def remove_student(roll):
    if roll in students:
        del students[roll]
        print("Student removed successfully.")
    else:
        print("Student not found.")


def add_course(roll, course):
    if roll in students:
        students[roll]["courses"].add(course)
        print(f" Course '{course}' added for student {roll}.")
    else:
        print("Student not found.")


def remove_course(roll, course):
    if roll in students:
        students[roll]["courses"].discard(course)
        print(f"Course '{course}' removed for student {roll}.")
    else:
        print("Student not found.")


def display_student(roll):
    if roll in students:
        data = students[roll]
        print("\n-- Student Details --")
        print(f"Roll No : {roll}")
        print(f"Name    : {data['name']}")
        print(f"Age     : {data['age']}")
        print(f"GPA     : {data['gpa']}")
        print(f"Courses : {', '.join(data['courses'])}")
        print("--------------------------\n")
    else:
        print("Student not found.")

# Example Usage (Matches your requirement)

# Add a new student
add_student(104, "John", 23, 3.6, ["Math", "Science"])

# Add another student
add_student(105, "Riya", 21, 3.7, ["English", "Math"])

# Modify courses
add_course(104, "History")
remove_course(105, "Math")

# Display students
display_student(104)
display_student(105)

# Remove a student
remove_student(105)

# Try to display removed student
display_student(105)
```
