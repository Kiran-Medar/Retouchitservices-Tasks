```python
students = []
grades = []
def add_student(name, grade):
    students.append(name)
    grades.append(grade)
    print(f"Added: {name} with grade {grade}")

def display_students():
    print("\n Student List ")
    for i in range(len(students)):
        print(f"{students[i]} : {grades[i]}")

def get_grade(name):
    if name in students:
        index = students.index(name)
        print(f"{name}'s grade is: {grades[index]}")
    else:
        print("Student not found!")

def average_grade():
    if grades:
        avg = sum(grades) / len(grades)
        print(f"Class Average Grade: {avg:.2f}")
    else:
        print("No grades to calculate average.")
        
add_student("John", 85)
add_student("Riya", 92)
add_student("George", 76)

display_students()
get_grade("Riya")
average_grade()
```

```python
#functions
marks_list = []

# 1. Function with parameter
def show_list(mylist):
    print("Function with parameter :", mylist)

# 2. Function without parameter (uses global list)
def display_list():
    print("Function without parameter:", marks_list)

# 3. Local vs Global variables
def local_example():
    local_list = [40, 50, 60]   # local variable
    print("Local list inside function :", local_list)

def global_example():
    global marks_list   # explicitly modify global list
    marks_list.append(100)
    print("Global list modified inside function :", marks_list)

# 4. Nested function with return
def process_list(lst):
    def calculate_stats(inner_list):   # nested function
        total = sum(inner_list)
        average = total / len(inner_list)
        highest = max(inner_list)
        lowest = min(inner_list)
        return {"total": total, "average": average, "highest": highest, "lowest": lowest}
    return calculate_stats(lst)

# Main program execution

n = int(input("How many marks do you want to enter? "))
for i in range(n):
    mark = int(input(f"Enter mark {i+1}: "))
    marks_list.append(mark)

print("\nInitial global marks list:", marks_list)

# Step 2: Call function with parameter
show_list([10, 20, 30])

# Step 3: Call function without parameter
display_list()

# Step 4: Local vs Global demonstration
local_example()
global_example()
print("Global list outside function ->", marks_list)

# Step 5: Nested function with return
results = process_list(marks_list)
print("Nested function result (marks stats) ->", results)
        
```
