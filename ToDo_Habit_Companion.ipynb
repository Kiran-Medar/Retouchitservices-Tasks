```python
tasks = []   

habits = {
    "Exercise": {"today": 0, "streak": 0},
    "Sleep Hours": {"today": 0, "streak": 0},
    "Drink Water": {"today": 0, "streak": 0}
}

def add_task():
    task_name = input("Enter a new task: ").strip()
    if task_name:
        tasks.append({"task": task_name, "done": False})
        print("Task added!\n")
    else:
        print("Task name cannot be empty.\n")

def view_tasks():
    print("\n TASK LIST ")
    if not tasks:
        print("No tasks available.\n")
        return

    for index, task in enumerate(tasks, start=1):
        status = "Right" if task["done"] else "Wrong"
        print(f"{index}. {task['task']}  [{status}]")
    print()

def complete_task():
    view_tasks()
    if not tasks:
        return

    try:
        num = int(input("Enter task number to mark done: "))
        if 1 <= num <= len(tasks):
            tasks[num - 1]["done"] = True
            print("Task marked as completed!\n")
        else:
            print("Invalid task number.\n")
    except ValueError:
        print("Please enter a number.\n")

def view_habits():
    print("\n HABITS ")
    for name, info in habits.items():
        print(f"{name} - Today: {info['today']} | Streak: {info['streak']}")
    print()

def update_habit():
    print("\nWhich habit did you complete?")
    habit_list = list(habits.keys())

    for i, h in enumerate(habit_list, start=1):
        print(f"{i}. {h}")

    try:
        choice = int(input("Enter number: "))
        if 1 <= choice <= len(habit_list):
            habit_name = habit_list[choice - 1]
            habits[habit_name]["today"] += 1
            habits[habit_name]["streak"] += 1
            print(f"Updated {habit_name}!\n")
        else:
            print("Invalid option.\n")
    except ValueError:
        print("Please enter a number.\n")

def daily_insight():
    print("\n DAILY WELLNESS INSIGHT ")

    if tasks:
        done = sum(1 for t in tasks if t["done"])
        task_score = (done / len(tasks)) * 100
    else:
        task_score = 0

    streak_total = sum(info["streak"] for info in habits.values())
    habit_score = min(streak_total * 15, 100)

    wellness = (task_score * 0.5) + (habit_score * 0.5)

    print(f"Task Completion Score: {task_score:.1f}%")
    print(f"Habit Streak Score : {habit_score:.1f}%")
    print(f"Overall Wellness Score: {wellness:.1f}%")

    if wellness >= 80:
        print("Youâ€™re doing amazing Keep the momentum!")
    elif wellness >= 50:
        print("Good job A little more consistency will help.")
    else:
        print("Slow day You can bounce back tomorrow!")

    print()

def menu():
    while True:
        print("   SMART TO-DO & HABIT COMPANION")
        print("1. Add Task")
        print("2. View Tasks")
        print("3. Mark Task as Completed")
        print("4. View Habits")
        print("5. Update Habit")
        print("6. Daily Wellness Insight")
        print("7. Exit")

        choice = input("Choose an option: ").strip()

        if choice == "1":
            add_task()
        elif choice == "2":
            view_tasks()
        elif choice == "3":
            complete_task()
        elif choice == "4":
            view_habits()
        elif choice == "5":
            update_habit()
        elif choice == "6":
            daily_insight()
        elif choice == "7":
            print("Goodbye! Stay healthy and productive!")
            break
        else:
            print("Invalid choice. Try again.\n")

menu()
```
